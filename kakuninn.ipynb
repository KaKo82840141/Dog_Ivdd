```python
# import tensorflow as tf
# import os

# # â˜…â˜… å¿…ãšã€æ–°ã—ãå­¦ç¿’ã•ã›ãŸãƒ¢ãƒ‡ãƒ«ã®ãƒ‘ã‚¹ã«ã—ã¦ãã ã•ã„ â˜…â˜…
# # ã‚‚ã—å‰ã®æ‰‹é †ã§ 'model_new.h5' ã«ãƒªãƒãƒ¼ãƒ ã—ã¦ã„ãŸã‚‰ã€ãã®åå‰ã«ã—ã¦ãã ã•ã„ã€‚
# MODEL_PATH = r"C:\Users\admin\Downloads\RNN-for-Human-Activity-Recognition-using-2D-Pose-Input-master\weights\model.h5"

# try:
#     print(f"--- ãƒ¢ãƒ‡ãƒ«ã®èª­ã¿è¾¼ã¿ã‚’è©¦ã—ã¾ã™ ---")
#     print(f"ãƒ‘ã‚¹: {MODEL_PATH}")

#     # ãƒ•ã‚¡ã‚¤ãƒ«ã®å­˜åœ¨ç¢ºèª
#     if not os.path.exists(MODEL_PATH):
#         print("\nâŒ ã‚¨ãƒ©ãƒ¼: æŒ‡å®šã•ã‚ŒãŸãƒ‘ã‚¹ã«ãƒ•ã‚¡ã‚¤ãƒ«ãŒè¦‹ã¤ã‹ã‚Šã¾ã›ã‚“ã€‚")
#     else:
#         model = tf.keras.models.load_model(MODEL_PATH)
#         print("\nğŸ‰ æˆåŠŸï¼ãƒ¢ãƒ‡ãƒ«ã¯æ­£å¸¸ã«èª­ã¿è¾¼ã‚ã¾ã—ãŸã€‚")
#         print("ã“ã‚Œã¯æ–°ã—ãå­¦ç¿’ã•ã›ãŸã€TensorFlow 2.xå¯¾å¿œã®ãƒ¢ãƒ‡ãƒ«ã§ã™ã€‚")
#         print("\nãƒ¢ãƒ‡ãƒ«ã®æ¦‚è¦:")
#         model.summary()

# except Exception as e:
#     print(f"\nâŒ ã‚¨ãƒ©ãƒ¼ãŒç™ºç”Ÿã—ã¾ã—ãŸã€‚")
#     print("ã‚¨ãƒ©ãƒ¼å†…å®¹:", e)
#     print("\nã“ã®ã‚¨ãƒ©ãƒ¼ãƒ¡ãƒƒã‚»ãƒ¼ã‚¸ãŒå‡ºãŸå ´åˆã€æŒ‡å®šã•ã‚ŒãŸãƒ‘ã‚¹ã®ãƒ•ã‚¡ã‚¤ãƒ«ãŒå¤ã„ãƒ¢ãƒ‡ãƒ«ã®ã¾ã¾ã§ã‚ã‚‹å¯èƒ½æ€§ãŒéå¸¸ã«é«˜ã„ã§ã™ã€‚")

import tensorflow as tf
import os
import numpy as np # n_inputãªã©ã®å®šç¾©ã«å¿…è¦

# --- ãƒ¢ãƒ‡ãƒ«ã®å®šç¾©ã‚’ã“ã“ã«å†ç¾ ---
# ä¿å­˜ã—ãŸãƒ¢ãƒ‡ãƒ«ã¨åŒã˜æ§‹é€ ã®ã‚¯ãƒ©ã‚¹ã‚’å®šç¾©ã™ã‚‹å¿…è¦ãŒã‚ã‚Šã¾ã™ã€‚
n_steps = 30
n_input = 84 # X_train[0][0] ã®é•·ã•ã€‚å®Ÿéš›ã®å€¤ã«åˆã‚ã›ã¦ãã ã•ã„ã€‚
n_hidden = 34
n_classes = 4

class LSTM_RNN(tf.keras.Model):
    def __init__(self, n_input, n_hidden, n_classes):
        super(LSTM_RNN, self).__init__()
        # TimeDistributedã¯Denseãƒ¬ã‚¤ãƒ¤ãƒ¼ã‚’å†…éƒ¨ã«æŒã¤ãŸã‚ã€
        # ã“ã®ã‚ˆã†ã«å†…å´ã§å®šç¾©ã™ã‚‹ã®ãŒä¸€èˆ¬çš„ã§ã™ã€‚
        self.time_dist_dense = tf.keras.layers.TimeDistributed(
            tf.keras.layers.Dense(n_hidden, activation='relu')
        )
        self.lstm1 = tf.keras.layers.LSTM(n_hidden, return_sequences=True)
        self.lstm2 = tf.keras.layers.LSTM(n_hidden)
        self.out = tf.keras.layers.Dense(n_classes)

    def call(self, x):
        x = self.time_dist_dense(x)
        x = self.lstm1(x)
        x = self.lstm2(x)
        x = self.out(x)
        return x

# --- ãƒ¢ãƒ‡ãƒ«ã®èª­ã¿è¾¼ã¿ ---
MODEL_PATH = r"C:\Users\admin\Downloads\RNN-for-Human-Activity-Recognition-using-2D-Pose-Input-master\weights\model.h5"

try:
    print(f"--- ãƒ¢ãƒ‡ãƒ«ã®èª­ã¿è¾¼ã¿ã‚’è©¦ã—ã¾ã™ ---")
    print(f"ãƒ‘ã‚¹: {MODEL_PATH}")

    if not os.path.exists(MODEL_PATH):
        print("\nâŒ ã‚¨ãƒ©ãƒ¼: æŒ‡å®šã•ã‚ŒãŸãƒ‘ã‚¹ã«ãƒ•ã‚¡ã‚¤ãƒ«ãŒè¦‹ã¤ã‹ã‚Šã¾ã›ã‚“ã€‚")
    else:
        # 1. ã¾ãšã€ã‚¯ãƒ©ã‚¹ã‹ã‚‰ãƒ¢ãƒ‡ãƒ«ã®ã€Œå™¨ã€ã‚’ä½œæˆã—ã¾ã™
        model = LSTM_RNN(n_input, n_hidden, n_classes)

        # 2. æ¬¡ã«ã€ãã®å™¨ã«ä¿å­˜ã—ãŸã€Œé‡ã¿ã€ã‚’èª­ã¿è¾¼ã¿ã¾ã™
        # â˜…â˜…â˜… load_model ã§ã¯ãªã load_weights ã‚’ä½¿ã„ã¾ã™ â˜…â˜…â˜…
        model.load_weights(MODEL_PATH)

        # ãƒ€ãƒŸãƒ¼ãƒ‡ãƒ¼ã‚¿ã§ä¸€åº¦ãƒ¢ãƒ‡ãƒ«ã‚’å‹•ã‹ã—ã€summary()ã‚’è¡¨ç¤ºå¯èƒ½ã«ã™ã‚‹
        dummy_input = np.zeros((1, n_steps, n_input), dtype=np.float32)
        model(dummy_input)

        print("\nğŸ‰ æˆåŠŸï¼ãƒ¢ãƒ‡ãƒ«ã®é‡ã¿ã¯æ­£å¸¸ã«èª­ã¿è¾¼ã‚ã¾ã—ãŸã€‚")
        print("ã“ã‚Œã¯ã‚ãªãŸã®å­¦ç¿’æ¸ˆã¿TensorFlow 2.xå¯¾å¿œãƒ¢ãƒ‡ãƒ«ã§ã™ã€‚")
        print("\nãƒ¢ãƒ‡ãƒ«ã®æ¦‚è¦:")
        model.summary()

except Exception as e:
    print(f"\nâŒ ã‚¨ãƒ©ãƒ¼ãŒç™ºç”Ÿã—ã¾ã—ãŸã€‚")
    print("ã‚¨ãƒ©ãƒ¼å†…å®¹:", e)
```
