```python
import pandas as pd
import numpy as np

# サンプルデータを作成
# 実際にはJSONから読み込んだデータを使用
data = {'left_wrist_x': [100, np.nan, np.nan, np.nan, np.nan, 200],
        'left_wrist_y': [200, np.nan, np.nan, np.nan, np.nan, 280]}
df = pd.DataFrame(data)

# 'linear'（線形）を指定して補間
# inplace=True はデータフレームを直接書き換えるオプション
df.interpolate(method='linear', inplace=True)

print(df)
```
